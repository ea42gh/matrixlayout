# Eigen/SVD Tables (matrixlayout.eigproblem)

This notebook introduces eigen/SVD table renderers.


## `render_eig_tex`
Emit LaTeX for eigen/SVD tables.


```python
from matrixlayout.eigproblem import render_eig_tex

spec = {
    "lambda": [2, 3],
    "ma": [1, 1],
    "evecs": [[[1, 0]], [[0, 1]]],
}
tex = render_eig_tex(spec, case="S")
tex[:200]
```

## `render_eig_svg`
Render eigen/SVD tables to SVG.


```python
from matrixlayout.eigproblem import render_eig_svg

spec = {
    "lambda": [2, 3],
    "ma": [1, 1],
    "evecs": [[[1, 0]], [[0, 1]]],
}
svg = render_eig_svg(spec, case="S")
svg[:200]
```
