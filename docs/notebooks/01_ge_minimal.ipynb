# GE minimal
Render a small GE grid as SVG and inspect the TeX.

Expected output: a single 2×2 matrix with standard delimiters.

```python
import sympy as sym
from IPython.display import SVG, display
from matrixlayout.ge import ge_grid_tex, ge_grid_svg
```

```python
matrices = [[None, sym.Matrix([[1, 2], [3, 4]])]]
tex = ge_grid_tex(matrices=matrices)

svg = ge_grid_svg(
    matrices=matrices,
    output_dir="./_out",
    output_stem="ge_min",
)
display(SVG(svg))
```

## Matrix product grid
A 2×2 example laid out as a block grid: [[None, B], [A, A@B]].

Expected output: a 2×2 block grid with B in the top-right and A·B in the bottom-right.

```python
A = sym.Matrix([[1, 2], [0, 1]])
B = sym.Matrix([[2, 0], [3, 1]])
matrices = [[None, B], [A, A * B]]

svg = ge_grid_svg(matrices=matrices)
display(SVG(svg))
```

## Troubleshooting
If rendering fails, try calling `ge_grid_tex` to inspect the TeX, or set `output_dir`
to capture artifacts for inspection.
