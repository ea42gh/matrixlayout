# Eigen/SVD table
Render an eigenproblem table as SVG without relying on la_figures.

Expected output: a compact eigenvalue table with two eigenpairs.

```python
from IPython.display import SVG, display
from matrixlayout import eigproblem_svg
```

```python
spec = {
    "lambda": [2, 3],
    "ma": [1, 1],
    "evecs": [[[1, 0]], [[0, 1]]],
}

svg = eigproblem_svg(spec, case="S")
display(SVG(svg))
```

## Troubleshooting
If rendering fails, switch to `eigproblem_tex` and inspect the TeX.
